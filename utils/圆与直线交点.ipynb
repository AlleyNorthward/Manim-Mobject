```python
def get_line_circle_intersection(line:Line, circle:Circle):
    vec1 = line.get_vector()
    vec2 = line.get_start() - circle.get_center()

    d = cross2d(vec1, vec2)/line.get_length()
    line_intersection()
    sin_alpha = d/circle.radius
    _alpha = np.arcsin(sin_alpha)
    _phi = line.get_angle()
    _theta1 = _alpha + _phi

    _theta2 = _theta1 + PI - 2*_alpha
    return (circle.point_at_angle(_theta1), circle.point_at_angle(_theta2))
```

```python
class Test(Scene):
    def construct(self):
        circle = Circle().scale(2.3)
        dot1 = Dot(circle.point_from_proportion(0.15))
        dot2 = Dot(circle.point_from_proportion(0.45))
        line = Line(dot1, dot2).set_length(20).set_stroke(width = 2)
        center = Dot(circle.get_center())
        dot3 = Dot(circle.point_from_proportion(0))
        x_line = Line(center, dot3).set_length(20).set_stroke(width=2)

        dot4 = Dot(midpoint(dot1.get_center(), dot2.get_center()))
        dashed_line1 = DashedLine(center.get_center(), dot4.get_center())
        dashed_line2 = DashedLine(center.get_center(), dot1.get_center())
        
        _phi_angle = Angle(line, x_line, quadrant=(-1,1), other_angle=True, radius=1.5)
        _phi_tex = MathTex(r"\phi").move_to(_phi_angle.get_center()).scale(0.6).shift(RIGHT*0.3+UP*0.1)

        _alpha_angle = Angle(line, dashed_line2, radius=0.8, quadrant=(1,-1))
        _alpha_tex = MathTex(r"\alpha").move_to(_alpha_angle.get_center()).scale(0.6).shift(LEFT*0.3+DOWN*0.2)

        _theta_angle = Angle(x_line, dashed_line2, radius = 0.8)
        _theta_tex = MathTex(r"\theta").move_to(_theta_angle.get_center()).scale(0.8).shift(UP*0.2+RIGHT*0.4)
        self.add(circle, dot1, dot2, line, center, x_line, dot4, dashed_line1, dashed_line2, _phi_angle, _phi_tex, _alpha_angle, _alpha_tex, _theta_angle, _theta_tex)

%manim $_RF Test
```
